```python
#append
fruits = ["apple","banana"]
fruits.append("mango")
print(fruits)
```

```python
#insert
fruits = [ "apple","banana"]
fruits.insert(1,"mango")
print(fruits)
```

```python
#remove
fruits = ["apple","banana","mango"]
fruits.remove("banana")
print(fruits)
```

```python
#pop()
Number = [10,20,30,40]
Number.pop()
print(Number)
```
